```python
# Final project of Python : Inventory management system

# Inventory dictionary
inventory = {}

def add_item(item_id, name, quantity, price):
    """This adds a new item into the inventory."""
    if item_id in inventory:
        print("Item with this ID already exists.")
    else:
        inventory[item_id] = {
            'name': name,
            'quantity': quantity,
            'price': price
        }
        print(f"Item '{name}' has been added with quantity :{quantity} and price: {price} ")

def update_item(item_id, quantity=None, price=None):
    """This Updates the quantity and price of an already existing inventory item."""
    if item_id in inventory:
        if quantity is not None:
            inventory[item_id]['quantity'] += quantity
        if price is not None:
            inventory[item_id]['price'] = price
        print(f"Item {item_id} updated successfully with quantity: {quantity} & price: {price}.")
    else:
        print("Item not found.")

def delete_item(item_id):
    """this deletes the entered item from the inventory."""
    if item_id in inventory:
        del inventory[item_id]
        print(f"Item:  {item_id}is deleted successfully from the inventory.")
    else:
        print("Item not found / consumed completely.")

def view_item(item_id):
    """This shows the details of the item we have asked for."""
    if item_id in inventory:
        item = inventory[item_id]
        print(f"ID: {item_id}, Item Name: {item['name']}, Quantity: {item['quantity']}, Price: ${item['price']:.2f}")
    else:
        print("Item is NOT found / consumed completely.")

def view_all_items():
    """View all items in the inventory."""
    if inventory:
        for item_id, item in inventory.items():
            print(f"ID: {item_id}, Name: {item['name']}, Quantity: {item['quantity']}, Price: ${item['price']:.2f}")
    else:
        print("No items in inventory OR All items are consumed from inventory.")

def main():
    while True:
        print("\n Inventory Management System")
        print("Enter the number as follows to perfom the operation:")
        print("1. Add Item in the inventory")
        print("2. Update an existing Item in inventory")
        print("3. Delete an already existing Item from inventory ")
        print("4. View record of a signle Item in the inventory")
        print("5. View All Items in the inventory")
        print("6. Exit from the I.M.S.")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            item_id = input("Enter the item ID: ")
            name = input("Enter the name of the item: ")
            quantity = int(input("Enter quantity: "))
            price = float(input("Enter price: "))
            add_item(item_id, name, quantity, price)

        elif choice == '2':
            item_id = input("Enter the item ID: ")
            quantity = input("Enter quantity to add or remove (or leave blank of no change): ")
            price = input("Enter new price (or leave blank if no change in price): ")

            quantity = int(quantity) if quantity else None
            price = float(price) if price else None
            update_item(item_id, quantity, price)

        elif choice == '3':
            item_id = input("Enter item ID to delete the item: ")
            delete_item(item_id)

        elif choice == '4':
            item_id = input("Enter item ID to view the the rocord: ")
            view_item(item_id)

        elif choice == '5':
            view_all_items()

        elif choice == '6':
            print("Exiting the system.")
            break

        else:
            print("Invalid choice entered. Enter the choice again pelase.")

if __name__ == "__main__":
    main()
```
